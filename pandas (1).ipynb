## An example: visualizing data in the notebook ✨

```python
from matplotlib import pyplot as plt
import numpy as np

# Generate 100 random data points along 3 dimensions
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# Map each onto a scatterplot we'll create with Matplotlib
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Some random data, created with JupyterLab!")
plt.show()
```

```python
#create dataframe
import pandas as pd
data={'Company':['GooG','Amazon','Cognizant','FB','FB'],
      'person':['Sam','Charlie','Car','Any','Sarah'],
      'Sales':[200,120,234,673,177]}
labels=['A','B','C','D','E']
df=pd.DataFrame(data=data,index=labels)
df
```

```python
byComp=df.groupby('Company')
print("\nGrouping by' Company' column and listing mean sales\n",'-'*34,sep='')
print(byComp.mean())
print("\nGrouping by' Company' column and listing sum sales\n",'-'*34,sep='')
print(byComp.sum())
#Note  dataframe conversion of the series and transpose
print("\nAll in one line of command (Stats for 'FB')\n",'-'*54,sep='')
print(pd.DataFrame(df.groupby('Company').describe().loc['FB']).transpose())
```

```python
import pandas as pd
df1=pd.DataFrame( {'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3']
                   'D':['D0','D1','D2','D3'],
                 index=[0,1,2,3])
df2=pd.DataFrame( {'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                 index=[4,5,6,7])
df3=pd.DataFrame( {'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                 index=[8,9,10,11])
print("\nThe DataFrame number 1\n",'-'*32,sep='-')
print(df1)
print("\nThe DataFrame number 2\n",'-'*32,sep='-')
print(df2)
print("\nThe DataFrame number 3\n",'-'*32,sep='-')
print(df3)

```

```python
df_cat1 = pd.concat([df1,df2,df3],axis=0)
print("\nAfter concatination along a row\n",'-'*54,sep='')
print(df_cat1)
```

```python
df_cat2 = pd.concat([df1,df2,df3],axis=1)
print("\nAfter concatination along a coumn\n",'-'*54,sep='')
print(df_cat2)
```

```python
df_cat2.fillna(value=0,inplace=True)
print("\nAfter filling missing value with zero\n",'-'*54,sep='')
df_cat2
```

```python
#Merging by a common "key"
#The merge function allows you to merge DataFrame together using similar logic as merging SQL table together..
```

```python
left=pd.DataFrame( {'Key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'Key':['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
print("\nThe DataFrame'left'\n",'-'*43,sep='')
print(left)
print("\nThe DataFrame'right'\n",'-'*43,sep='')
right
                   
```

```python
merge1 = pd.merge(left,right,how='inner',on='Key')
print("\nAfter simple merging with 'inner' method\n",'-'*43,sep='')
print(merge1)
```

```python
#Merging on a set of keys
```

```python
left = pd.DataFrame({'Key1':['K0','K1','K2','K3'],
                     'Key2':['K0','K1','K2','K3'],
                        'A':['A0','A1','A2','A3'],
                        'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'Key1':['K0','K1','K2','K3'],
                      'Key2':['K0','K1','K2','K3'],
                         'C':['C0','C1','C2','C3'],
                         'D':['D0','D1','D2','D3']})
merge1=pd.merge(left,right,on=['Key1','Key2'])
merge1
```

```python
merge1=pd.merge(left,right,how='inner',on=['Key1','Key2'])
merge1
```

```python
merge1=pd.merge(left,right,how='right',on=['Key1','Key2'])
merge1
```

```python
merge1=pd.merge(left,right,how='left',on=['Key1','Key2'])
merge1
```
